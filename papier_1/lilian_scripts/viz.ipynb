```python
import pandas as pd
import numpy as np
from tqdm import tqdm
import plotly.express as px
```

```python
distances_in = pd.read_csv('data/in_distances.csv').drop(columns = ['Unnamed: 0'])
distances_out = pd.read_csv('data/out_distances.csv').drop(columns = ['Unnamed: 0', 'Unnamed: 0.1'])
```

```python
distances_out.index = range(len(distances_in), len(distances_in) + len(distances_out))
```

```python
distances = pd.concat([distances_in, distances_out])
distances['origin'] = ['in'] * len(distances_in) + ['out'] * len(distances_out)
```

```python
distances.columns
```

```python
px.histogram(
                distances[distances['MSP'] < .0006], 
                x = 'MSP',
                color = 'origin',
                template = 'none',
                nbins = 100
            )
```

```python
px.histogram(
                distances[distances['E'] > 1.313], 
                x = 'E',
                color = 'origin',
                template = 'none',
                nbins = 100
            )
```

```python
px.histogram(
                distances, 
                x = 'D_M',
                color = 'origin',
                template = 'none',
                nbins = 100
            )
```
